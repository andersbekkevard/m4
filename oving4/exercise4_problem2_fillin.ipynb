# Fix point method

```python
import numpy as np
import matplotlib.pyplot as plt

r = 0.6823278038280193 # error ~ 1e-16
```

```python
def FixpointIteration(g, x0, error=1e-6, max_iter=100, verbose=False):
	x = x0
	for k in range(max_iter):
		x_new = g(x)
		if np.abs(x_new - x) < error:
			return x_new, k
		x = x_new
	return x, max_iter
```

### $g_1(x)=1-x^3$

```python
def g(x): 
    return 1 - x**3

def g_derivative(x):
    return -3 * x**2

print("g(r) =", g(r))
print("r =", r)
print("verified:", np.allclose(g(r), r))

x0 = r - 0.1
print(f"{'Iteration':>10} | {'x':>20}")
print("-" * 33)
x = x0
for i in range(3):
    print(f"{i:10d} | {x:20.12f}")
    x = g(x)

delta = 0.5
for x in np.linspace(r - delta, r + delta, 3):
    if np.abs(g_derivative(x)) < 1:
        print(f"x = {x} is a good starting point")
    else:
        print(f"x = {x} is not a good starting point")
print("g_derivative(r) =", g_derivative(r))

x = np.linspace(r - delta, r + delta, 100)
y = [FixpointIteration(g, xx, 1e-6, 1000) for xx in x]
fig, ax = plt.subplots(1,3,figsize=(15,4))
ax[0].plot(x,[yy[0] for yy in y],'o-',markersize=2)
ax[0].set_xlabel('$x$')
ax[0].set_ylabel('Converged iteration point')
ax[0].set_title('Fixed Point Iteration')
ax[1].plot(x,np.abs([yy[0]-r for yy in y]),'o-',markersize=2)
ax[1].set_xlabel('$x$')
ax[1].set_ylabel('Error to root')
ax[1].set_title('Error to root')
ax[2].plot(x,[yy[1] for yy in y],'o-',markersize=2)
ax[2].set_xlabel('$x$')
ax[2].set_ylabel('Number of iterations needed')
ax[2].set_title('Number of iterations to converge')
plt.show()
```

### $g_5(x)=x-\frac{x^3+x-1}{3x^2+1}$

```python
def g(x):
    return x - (x**3 + x - 1.0) / (3.0 * x**2 + 1.0)

def g_derivative(x):
    f = x**3 + x - 1.0
    fp = 3.0 * x**2 + 1.0
    fpp = 6.0 * x
    return (f * fpp) / (fp**2)

print("g(r) =", g(r))
print("r =", r)
print("verified:", np.allclose(g(r), r))

x0 = r - 0.1
print(f"{'Iteration':>10} | {'x':>20}")
print("-" * 33)
x = x0
for i in range(3):
    print(f"{i:10d} | {x:20.12f}")
    x = g(x)

delta = 0.5
for x in np.linspace(r - delta, r + delta, 3):
    if np.abs(g_derivative(x)) < 1:
        print(f"x = {x} is a good starting point")
    else:
        print(f"x = {x} is not a good starting point")
print("g_derivative(r) =", g_derivative(r))

x = np.linspace(r - delta, r + delta, 100)
y = [FixpointIteration(g, xx, 1e-6, 1000) for xx in x]
fig, ax = plt.subplots(1,3,figsize=(15,4))
ax[0].plot(x,[yy[0] for yy in y],'o-',markersize=2)
ax[0].set_xlabel('$x$')
ax[0].set_ylabel('Converged iteration point')
ax[0].set_title('Fixed Point Iteration')
ax[1].plot(x,np.abs([yy[0]-r for yy in y]),'o-',markersize=2)
ax[1].set_xlabel('$x$')
ax[1].set_ylabel('Error to root')
ax[1].set_title('Error to root')
ax[2].plot(x,[yy[1] for yy in y],'o-',markersize=2)
ax[2].set_xlabel('$x$')
ax[2].set_ylabel('Number of iterations needed')
ax[2].set_title('Number of iterations to converge')
plt.show()
```
