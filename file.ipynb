```python
def f(x):
	return x**2
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create x values
x = np.linspace(-10, 10, 100)
y = f(x)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'b-', linewidth=2)
plt.grid(True, alpha=0.3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('f(x) = x²')
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()
```
